**TUGAS 3 VISUALISASI DATA DAN INFORMASI**

1. Data yang Digunakan untuk Visualisasi:

```python
import pandas as pd

#Read Dataset
data = pd.read_csv('/content/diabetes.csv')
data.head(10)
```

Dataset ini berisi informasi kesehatan terkait diabetes. Saya akan memilih untuk memvisualisasikan distribusi kadar Glukosa berdasarkan Outcome (Case atau Control), yang menunjukkan apakah individu memiliki diabetes atau tidak.

2. Tujuan, User, Fungsi, dan Tone:

Tujuan: Menampilkan distribusi kadar glukosa antara individu yang terdiagnosis diabetes (Case) dan yang tidak (Control).

User: Dokter, peneliti kesehatan, atau profesional medis yang ingin melihat pola distribusi kadar glukosa antara pasien dengan dan tanpa diabetes.

Fungsi: Membantu pengguna dalam memahami bagaimana distribusi glukosa bervariasi antara kelompok yang terdiagnosis diabetes dan yang tidak.

Tone: Profesional dan informatif. Visualisasi harus memberikan wawasan yang mudah dipahami.

3. Visualisasi dengan Python (Seaborn dan Matplotlib):

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Mengatur gaya visualisasi Seaborn
sns.set(style="whitegrid")

# Membuat plot boxplot untuk melihat distribusi glukosa berdasarkan Outcome (Case dan Control)
plt.figure(figsize=(10, 6))
sns.boxplot(x='Outcome', y='Glucose', data=data)

# Menambah judul dan label
plt.title('Distribusi Glukosa Berdasarkan Outcome (Diabetes/Non-Diabetes)', fontsize=16)
plt.xlabel('Outcome', fontsize=12)
plt.ylabel('Kadar Glukosa', fontsize=12)

# Tampilkan plot
plt.tight_layout()
plt.show()
```

**Penjelasan Visualisasi:**

Grafik di atas adalah boxplot yang menunjukkan distribusi kadar glukosa berdasarkan dua kelompok Outcome:

- Case: Orang yang terdiagnosis diabetes.
- Control: Orang yang tidak terdiagnosis diabetes.

Interpretasi:
- Kelompok Case (diabetes) cenderung memiliki kadar glukosa yang lebih tinggi dibandingkan dengan kelompok Control (non-diabetes).
- Median kadar glukosa pada kelompok Case lebih tinggi dibandingkan kelompok Control.
- Ada beberapa outlier di kedua kelompok, terutama pada kelompok Control di bagian kadar glukosa yang lebih rendah.

**Berdasarkan data ini:**
Kadar glukosa lebih tinggi dihubungkan dengan kasus diabetes, yang sesuai dengan pengetahuan medis bahwa kadar glukosa tinggi merupakan indikator diabetes.
